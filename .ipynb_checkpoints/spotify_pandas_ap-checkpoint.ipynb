```python
#-------------------------------------------------
# # function to flatten json object

def flatten_json(y):
    out = {}
    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x
    flatten(y)
    return out
# ---------------------------------------------------  
    
```

```python
def averages_summary(raw_data, agg_type):

    avg_danceability = df_clean["danceability"].agg(agg_type)
    avg_energy = df_clean["energy"].agg(agg_type)
    avg_key = df_clean["key"].agg(agg_type)
    avg_loudness = df_clean["loudness"].agg(agg_type)
    avg_mode = df_clean["mode"].agg(agg_type)
    avg_speechiness = df_clean["speechiness"].agg(agg_type)
    avg_acousticness = df_clean["acousticness"].agg(agg_type)
    avg_instrumentalness = df_clean["instrumentalness"].agg(agg_type)
    avg_liveness = df_clean["liveness"].agg(agg_type)
    avg_tempo = df_clean["tempo"].agg(agg_type)
    avg_valence = df_clean["valence"].agg(agg_type)
    avg_duration_ms = df_clean["duration_ms"].agg(agg_type)
    avg_time_signature = df_clean["time_signature"].agg(agg_type)

    summary_df = pd.DataFrame( [{'danceability': avg_danceability, 
                              'energy': avg_energy,
                              'key': avg_key,
                              'loudness': avg_loudness,
                              'mode': avg_mode,
                              'speechiness': avg_speechiness,
                              'acousticness': avg_acousticness,
                              'instrumentalness': avg_instrumentalness,
                              'liveness': avg_liveness,
                              'tempo': avg_tempo,
                              'valence': avg_valence,
                              'duration_ms': avg_duration_ms,
                              'time_signature': avg_time_signature}])

    print(summary_df)
```
